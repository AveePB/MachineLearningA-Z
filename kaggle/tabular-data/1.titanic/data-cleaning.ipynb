## Import libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

## Load data and drop useless features

```python
df_train = pd.read_csv("./csv/train.csv", index_col="PassengerId")
df_train.head()
```

```python
df_train.info()
```

```python
df_test = pd.read_csv("./csv/test.csv", index_col="PassengerId")
df_test.head()
```

```python
df_test.info()
```

```python
# Drop columns
useless_feats = ["Name", "Cabin", "Ticket"]
df_train = df_train.drop(columns=useless_feats)
df_test = df_test.drop(columns=useless_feats)

# Drop records 
df_train = df_train.loc[df_train["Embarked"].notna(), :]
df_test = df_test.loc[df_test["Fare"].notna(), :]
```

```python
df_train.info()
```

```python
df_test.info()
```

## Data formating and improving 

### Use median to replace age NaNs

```python
age_median = df_train["Age"].median()
male_age_median = df_train.loc[(df_train["Sex"] == "male"), "Age"].median()
female_age_median = df_train.loc[(df_train["Sex"] == "female"), "Age"].median()

df_train.loc[:, "Age"] = df_train["Age"].fillna(age_median)
df_test.loc[:, "Age"] = df_test["Age"].fillna(age_median)
```

```python
print("Male age median: ", male_age_median)
print("Female age median: ", female_age_median)
print("Age median: ", age_median, "\n")

print("Training data frame age nan count: ", df_train["Age"].isna().sum())
print("Test data frame age nan count: ", df_test["Age"].isna().sum())
```

### Encoding categorical features

```python
df_train.info()
```

```python
df_test.info()
```

```python
# Sex encoding
df_train["Sex"] = df_train["Sex"].map({"female": 0, "male": 1})
df_test["Sex"] = df_test["Sex"].map({"female": 0, "male": 1})

# Embarked encoding
df_train["Embarked"] = df_train["Embarked"].map({"C": 0, "Q": 1, "S": 2})
df_test["Embarked"] = df_test["Embarked"].map({"C": 0, "Q": 1, "S": 2})
```

```python
df_train.head()
```

```python
df_test.head()
```

### Scale continous and descrite numerical features

```python
df_train.describe()
```

```python
from sklearn.preprocessing import StandardScaler

# Age scaling
age_scaler = StandardScaler()
df_train["Age"] = age_scaler.fit_transform(df_train[["Age"]])
df_test["Age"] = age_scaler.transform(df_test[["Age"]])

# SibSp scaler
sibsp_scaler = StandardScaler()
df_train["SibSp"] = sibsp_scaler.fit_transform(df_train[["SibSp"]])
df_test["SibSp"] = sibsp_scaler.transform(df_test[["SibSp"]])

# Parch sclaer
parch_scaler = StandardScaler()
df_train["Parch"] = parch_scaler.fit_transform(df_train[["Parch"]])
df_test["Parch"] = parch_scaler.transform(df_test[["Parch"]])

fare_scaler = StandardScaler()
df_train["Fare"] = fare_scaler.fit_transform(df_train[["Fare"]])
df_test["Fare"] = fare_scaler.transform(df_test[["Fare"]])
```

## Save data to csv file

```python
df_train.to_csv("./csv/clean-train.csv")
df_test.to_csv("./csv/clean-test.csv")
```
