## Import libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

## Load and prepare data

```python
df_train = pd.read_csv("./csv/clean-train.csv", index_col="PassengerId")
df_train.head()
```

```python
df_test = pd.read_csv("./csv/clean-test.csv", index_col="PassengerId")
df_test.head()
```

```python
features = ["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]
```

```python
df_train = df_train.sample(n=len(df_train))
X_train = df_train[features].to_numpy()
y_train = df_train["Survived"].to_numpy()

X_test = df_test[features].to_numpy()
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

## Train ML models

### Decision tree

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

#### Save decision tree submission to csv

```python
y_pred = model.predict(X_test)

df_submission = pd.DataFrame({"Survived": y_pred})
df_submission.set_index(df_test.index).to_csv("./submissions/decision-tree.csv")
```

### Random forest

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

#### Save decision tree submission to csv

```python
y_pred = model.predict(X_test)

df_submission = pd.DataFrame({"Survived": y_pred})
df_submission.set_index(df_test.index).to_csv("./submissions/random-forest.csv")
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()
model.fit(X_train, y_train)
```

#### Save decision tree submission to csv

```python
y_pred = model.predict(X_test)

df_submission = pd.DataFrame({"Survived": y_pred})
df_submission.set_index(df_test.index).to_csv("./submissions/knn.csv")
```

### SVC

```python
from sklearn.svm import SVC

model = SVC()
model.fit(X_train, y_train)
```

#### Save decision tree submission to csv

```python
y_pred = model.predict(X_test)

df_submission = pd.DataFrame({"Survived": y_pred})
df_submission.set_index(df_test.index).to_csv("./submissions/svc.csv")
```

### Logistic regression

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

#### Save decision tree submission to csv

```python
y_pred = model.predict(X_test)

df_submission = pd.DataFrame({"Survived": y_pred})
df_submission.set_index(df_test.index).to_csv("./submissions/logistic-regression.csv")
```

### XGBoost classifier

```python
from xgboost.sklearn import XGBClassifier

model = XGBClassifier()
model.fit(X_train, y_train)
```

#### Save decision tree submission to csv

```python
y_pred = model.predict(X_test)

df_submission = pd.DataFrame({"Survived": y_pred})
df_submission.set_index(df_test.index).to_csv("./submissions/xgb-classifier.csv")
```
