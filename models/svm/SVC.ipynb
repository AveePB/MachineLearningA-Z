# Support Vector Classifier
This notebook has basic implementation of SVC algorithm for binary classification. It is used to visualize how models learn.

#### Import libraries and dataset

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np
```

#### Clean and split dataset

```python
# Load data
X, y = make_classification(n_samples=1_000, n_features=2, n_classes=2, n_redundant=0, random_state=12)
X0, X1 = [], []

for i, data in enumerate(X):
    if y[i] == 0:
        X0.append(data)
    else:
        X1.append(data)

# Convert to numpy
X0 = np.array(X0)
X1 = np.array(X1)

# Visualize
plt.scatter(X0[:, 0], X0[:, 1], c="blue", label="0")
plt.scatter(X1[:, 0], X1[:, 1], c="red", label="1")

plt.title("Generated data set")
plt.xlabel("First feature")
plt.ylabel("Second feature")
plt.legend()
plt.show()

print("Number of samples:", len(X))
print("Example sample:")
print("Data:", X[0])
print("Target:", y[0])
```

```python
# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=44)
print("Size of training dataset:", len(X_train))
print("Size of test dataset:", len(X_test))
```

```python
# Scale data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Example scaled sample:")
print("Data:", X_train_scaled[0])
```
