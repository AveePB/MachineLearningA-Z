<h2>Decision Tree Regression</h2>

<h4>How does it work?</h4>

- Splitting the Data: The algorithm starts at the root of the tree and splits the data based on the feature that results in the best possible split. This is typically done by minimizing the variance within the resulting subsets;

- Making Decisions: Each internal node of the tree represents a decision based on a specific feature and a threshold value. The data is divided into two subsets: one where the feature value is less than or equal to the threshold, and the other where it is greater;

- Recursive Splitting: This process of splitting and creating decision nodes continues recursively for each subset, further partitioning the data into smaller and smaller groups. The recursion continues until a stopping criterion is met, such as a maximum depth of the tree, a minimum number of samples in a node, or a minimum reduction in variance;

- Predicting Values: Once the tree is fully grown, predictions are made by traversing the tree from the root to a leaf node based on the feature values of the input data. The value in the leaf node is the predicted continuous value;

<h4>Key Characteristics</h4>

- Interpretability: Decision trees are easy to visualize and interpret;

- Non-linearity: Decision trees can capture non-linear relationships between features and the target variable;

- Prone to Overfitting: Without proper pruning or setting constraints, decision trees can become very complex and overfit the training data;

- Variance Reduction: The splitting criteria often aim to minimize the variance in the target variable within each subset;

$$
\text{Variance reduction} = \text{Variance before split} - \left( \frac{n_{\text{left}}}{n} \cdot \text{Variance}_{\text{left}} + \frac{n_{\text{right}}}{n} \cdot \text{Variance}_{\text{right}} \right) 
$$


<h4>Importing dataset</h4>

```python
import pandas as pd

df = pd.read_csv("Position_Salaries.csv")
X = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
```

<h4>Training model</h4>

```python
from sklearn.tree import DecisionTreeRegressor

dec_tree_reg = DecisionTreeRegressor()
dec_tree_reg.fit(X, y)
```

<h4>Visualising the decision tree regression results</h4>

```python
import matplotlib.pyplot as plt
import numpy as np 

X_grid = np.arange(min(X), max(X), 0.1).reshape(-1, 1)

plt.scatter(X, y, c = "red")
plt.plot(X_grid, dec_tree_reg.predict(X_grid),  c ="blue")
plt.title("Decision Tree Regression: Position v Salary")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```
